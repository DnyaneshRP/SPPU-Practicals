**2) Load the Dataset**

```python
# Load the NBA dataset
df1 = pd.read_csv("nba.csv")
```

**4) Check for missing values**

```python
# Check for missing values
df1.isnull().sum()
```

```python
# Percentage of missing values
df1.isnull().mean() * 100
```

**5) Handle Missing Values**

```python
df1['Number'] = df1['Number'].fillna(df1['Number'].mean())
df1['Age'] = df1['Age'].fillna(df1['Age'].mean())
df1['Weight'] = df1['Weight'].fillna(df1['Weight'].mean())
df1['Salary'] = df1['Salary'].fillna(df1['Salary'].mean())
```

```python
# Check for missing values
df1.isnull().sum()
```

**Group By**

**1) Group By Category: Height & Numerical: Weight**

```python
# Group the dataset by Height
height_groups = df1.groupby("Height")

# Display first 5 rows from each height group
height_groups.head()
```

```python
# Get statistical summary of Weight grouped by Height
height_groups["Weight"].describe()
```

**2) Group By Category: Team & Numerical: Salary**

```python
# Group the dataset by the 'Team' column
team_groups = df1.groupby("Team")

# Display the first 5 rows from each team group
team_groups.head()
```

```python
# Get statistical summary of Salary for each team
team_groups["Salary"].describe()
```

**3) Group By Category: Age Group (Created) & Numerical: Salary**

```python
# Define the bin ranges for age grouping
bins = [19, 25, 31, 36, 40]

# Define labels for each age group
labels = ["19-24", "25-30", "31-35", "36-40"]

# Create a new column 'AgeGroup' by categorizing Age into bins
df1["AgeGroup"] = pd.cut(df1["Age"], bins=bins, labels=labels)

# Display the updated dataset with the new AgeGroup column
df1
```

```python
# Count how many players fall into each AgeGroup category
df1["AgeGroup"].value_counts()
```

```python
# Group the dataset by the 'AgeGroup' column
# observed=True ensures only the groups that actually appear in the data are shown
age_groups = df1.groupby("AgeGroup", observed=True)
```

```python
# Get statistical summary of Salary for each AgeGroup
age_groups["Salary"].describe()
```
