<h3>2. Create a histogram for each feature in the dataset to illustrate the feature distributions.</h3>

**a) Histogram of feature: Sepal Length**

```python
sns.histplot(data=df, x="SepalLengthCm", bins=10, kde=True)
```

**Inference:**

- The distribution is slightly right-skewed.
- Most values lie between 5.0 cm and 6.5 cm.
- No extreme outliers are visible.

**b) Histogram of feature: Sepal Width**

```python
sns.histplot(data=df, x="SepalWidthCm", bins=10, kde=True)
```

**Inference:**

1. The distribution is moderately symmetric.
2. Most values are between 2.5 cm and 3.5 cm.
3. A few values fall outside the main range, indicating possible outliers.

**c) Histogram of feature: Petal Length**

```python
sns.histplot(data=df, x="PetalLengthCm", bins=10, kde=True)
```

**Inference:**

- The distribution is bimodal, showing two distinct peaks.
- This indicates clear separation among flower species.
- Values mostly range from 1.0 cm to 6.5 cm.

**d) Histogram of feature: Petal Width**

```python
sns.histplot(data=df, x="PetalWidthCm", bins=10, kde=True)
```

**Inference:**

- The distribution is also bimodal.
- Most values lie between 0.1 cm and 2.5 cm.
- This feature is highly useful for classification.

<h3>3. Create a boxplot for each feature in the dataset.</h3>

**a) Boxplot of All Numerical Features**

```python
sns.boxplot(data=df.drop(["Id", "Species"], axis=1))
```

**Inference:** 
- Sepal width shows outliers
- Petal features have wider spread and better separation.

<b>b) Boxplot of Sepal Length by Species</b>

```python
sns.boxplot(data=df, x="Species", y="SepalLengthCm")
```

**Inference:** 
- Virginica has the highest sepal length
- Versicolor in between
- Setosa has the lowest.

<b>c) Boxplot of Sepal Width by Species</b>

```python
sns.boxplot(data=df, x="Species", y="SepalWidthCm")
```

**Inference:**
- Setosa has the highest median sepal width.
- Versicolor and Virginica have similar, lower medians.
- A few outliers are present in virginica.

<b>d) Boxplot of Petal Length by Species</b>

```python
sns.boxplot(data=df, x="Species", y="PetalLengthCm")
```

**Inference:**
- Setosa has the smallest petal length with some outliers.
- Virginica has the largest petal length.
- Clear separation among all three species.

<b>e) Boxplot of Petal Width by Species</b>

```python
sns.boxplot(data=df, x="Species", y="PetalWidthCm")
```

**Inference:**
- Setosa has the smallest petal width with some outliers.
- Virginica has the largest petal width.
- Clear separation among species.

<h3>4. Compare Distributions and Identify Outliers</h3>

**Inference:**
- Sepal width shows outliers.
- Sepal length has moderate spread with no major outliers.
